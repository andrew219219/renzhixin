```python
def search_keyword(input_dataframe, key_word):
    for col_index, col_seria in input_dataframe.items():
        for row_index, row_value in col_seria.items():
            value_no_space = str(row_value).replace(' ', '')
            if value_no_space == key_word:
                return row_index, col_index
    print('find nothing!!!')
```

```python
def fetch_content_by_rkck(input_dataframe, row_keyword, column_keyword):
    row_index = search_keyword(input_dataframe, row_keyword)[0]
    column_index = search_keyword(input_dataframe, column_keyword)[1]
    print(row_index, column_index)
    return input_dataframe.iloc[row_index, column_index]
```
